```python
# Импорт модуля pandas
import pandas as pd
 
# Чтение csv-файла
df = pd.read_csv(io.BytesIO(uploaded['world-happiness-report.csv']))
df # Короткий вывод
```

```python
# Создаем список названий столбцов с критериями, которые будем потом подсчитывать.
# Так как первые 2 столбца содержат страну и год - удаляем их из списка
column_lst = list(df)
column_lst.remove('Country name')
column_lst.remove('year')

# Добавляем в dataframe столбец с подсчитанным средним значением всех критериев для каждой строки
df['HAPPINESS'] = df[column_lst].sum(axis=1)/9

# Делаем новый пустой dataframe для красивого вывода результата
result_df = pd.DataFrame()

# Для годов 2018-2020 находим страну с наибольшим "счастьем" граждан
for year in range(2018, 2021):
  # Выборка по году
  tmp_df = df.loc[df['year'].eq(year)]
  # Сортируем по убыванию выборку, берем первую строку и передаем в пустой dataframe result_df с сохранением индекса
  result_df = result_df.append(tmp_df.sort_values(by='HAPPINESS', ascending=False).head(1), ignore_index=False)

result_df # Короткий вывод
```

```python
# Импорт модуля для графиков
import plotly.graph_objects as go

# Выборка по году
year = 2018
pie_df = df.loc[df['year'].eq(year)]

# Создаем фигуру с диаграммой
fig = go.Figure(
    data=[go.Bar(x=pie_df['Country name'], y=pie_df['HAPPINESS'])], # Диаграмма со столбцами из pie_df
    layout_title_text="HAPPINESS for " + str(year) # Заголовок
)

fig.show() # Отрисовка
```
