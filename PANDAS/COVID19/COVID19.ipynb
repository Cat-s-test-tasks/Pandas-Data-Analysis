```python
# Загрузка локального csv-файла

from google.colab  import files
import io

uploaded = files.upload()
```

```python
# Импорт модуля pandas
import pandas as pd

# Чтение csv-файла
df = pd.read_csv(io.BytesIO(uploaded['Latest Covid-19 India Status.csv']))
df # Короткий вывод
```

```python
# Делаем новый dataframe для красивого вывода результата
result_df = df.copy()
result_df = result_df.drop(['Total Cases', 'Active', 'Discharged', 'Deaths', 'Active Ratio (%)'], axis = 1)

print('Топ 5 штатов по % выздоровления\n')
# Сортируем по убыванию выборку выздоровления, берем первые 5 строк
result_df.sort_values(by='Discharge Ratio (%)', ascending=False).head(5).drop(['Death Ratio (%)'], axis = 1) # Короткий вывод
```

```python
print('Топ 5 штатов по % смертности\n')
# Сортируем по убыванию выборку смертности, берем первые 5 строк
result_df.sort_values(by='Death Ratio (%)', ascending=False).head(5).drop(['Discharge Ratio (%)'], axis = 1) # Короткий вывод
```

```python
import plotly.graph_objects as go

# График соотношения смертей/лечения/выздоровления по штатам
fig = go.Figure(data=[
    go.Bar(name='Deaths', x=df['State/UTs'], y=df['Deaths']),
    go.Bar(name='Active', x=df['State/UTs'], y=df['Active']),
    go.Bar(name='Discharged', x=df['State/UTs'], y=df['Discharged'])
])

# Отображение
fig.update_layout(barmode='stack', height=1000)
fig.show()
```
